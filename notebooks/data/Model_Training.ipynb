```python
import pandas as pd
```

```python
df=pd.read_csv('final_data')
```

```python
d=df.copy()
```

```python
d.drop('ID',axis=1,inplace=True)
```

```python
d.head()
```

```python
X=d.drop('def_pay',axis=1)
y=d.def_pay
```

```python
d.columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.30,random_state=42)

X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.feature_selection import SelectKBest, chi2, f_classif
```

```python
scaler=StandardScaler()
```

```python

label_encoder = LabelEncoder()
X_encoded = X.apply(label_encoder.fit_transform)

selector_chi2 = SelectKBest(score_func=chi2, k=7)  # Select top 7 features using Chi-squared test
X_selected_chi2 = selector_chi2.fit_transform(X_encoded, y)

cols_idxs = selector_chi2.get_support(indices=True)
features_df_new =X_encoded.iloc[:,cols_idxs]

print(features_df_new)
key1=list(features_df_new.columns)
```

```python
# Perform feature selection using ANOVA test
selector_anova = SelectKBest(score_func=f_classif, k=7)  # Select top 7 features using ANOVA test
X_selected_anova = selector_anova.fit_transform(X_encoded, y)

cols_idxs = selector_anova.get_support(indices=True)
features_df_new1 =X_encoded.iloc[:,cols_idxs]

type(features_df_new1)

key=list(features_df_new1.columns)
```

```python
pd.DataFrame(X_selected_anova,columns=key).head()
```

```python
X_selected = pd.concat([pd.DataFrame(X_selected_chi2,columns=key1), pd.DataFrame(X_selected_anova,columns=key)], axis=1)
```

```python
X_selected.head()
X_selected.to_csv(r'C:\Users\shprs\Downloads\Stats\CreditCard Default\notebooks\data\imp_data')
```

```python
X_selected.head()
```

```python

X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

model = LogisticRegression()
model.fit(X_train_scaled, y_train)

y_pred = model.predict(X_test_scaled)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

print("Classification Report:")
print(classification_report(y_test, y_pred))

conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)
```
